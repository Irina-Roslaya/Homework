```python
import pandas as pd
url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/data_stat/yandex-stat-q.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
df["Фраза"]=df["Поисковая фраза"].str.lower()
df.head(5)
```

```python
l=list(df["Фраза"].unique())
m=[]
mm=[]
import re
for i in l:
    a=re.split(' | \| /',i)    
    for j in a:
        m.append(j)
new_m = list(dict.fromkeys(m))
for i in new_m:
    if "[" in i:
        i= i[:i.index('[')]+i[i.index('[')+1:]
    if ")" in i:
        i= i[:i.index(')')]+i[i.index(')')+1:]
    if "(" in i:
        i= i[:i.index('(')]+i[i.index('(')+1:]
    if "|" in i:
        i= i[:i.index('|')]+i[i.index('|')+1:]
    if "*" in i:
        i= i[:i.index('*')]+i[i.index('*')+1:]
    if "\z" in i:
        i= ''
    mm.append(i)
```

```python
vsc=[]
nu=[]
for i in mm:
    try:
        vs=df["Фраза"].str.contains(i).sum()
        nu.append(vs)
        vsc.append(f"Встречаемость слова {i}: {vs}")
    except:
        break
```

```python
nu=sorted(nu, reverse=True)
llll=[]
for i in nu:
    for j in vsc:
        if f"{i}" in j:
            llll.append(j)
```
