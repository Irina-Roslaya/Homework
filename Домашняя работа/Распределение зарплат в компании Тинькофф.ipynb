```python
import pandas as pd
import json
df_link = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/tinkoff_hh.json"
```

```python
with open('странный json.txt', encoding='utf8') as data:
    file = data.readlines()
```

```python
list1 = [json.loads(i) for i in file if 'index' not in i]
```

```python
df = pd.DataFrame(list1)
df.head(4)
```

```python
df2 = df
```

```python
def income_to_n(row):
    s = row['income'].split()
    if row['income'] == 'не указан':
        return 0
    elif 'до' in row['income']:
        return int(s[1])*1000
    else:
        return int(s[1])*1000
```

```python
df2['income'] = df2.apply(income_to_n, axis=1)
df2
```

```python
vacancys = df2.groupby(['vacancy'])['income'].mean().sort_values(ascending=False)
vacancys
```
