```python
import pandas as pd

url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/new_year_film.csv"
f=pd.read_csv(url)
f.head()
```


Задания
Определите средний рейтинг фильмов, представленных в данных.

```python
films=f[f.type=='фильм']
films.head()
```

```python
films_r=films['ranking'].replace({",":"."},regex=True)
```

```python
films_r.index
```

```python
endlist=[]
for i in films_r.index:
    m=0
    m=films_r[i]
    endlist.append(float(m))
endlist
```

```python
from statistics import mean
endw=mean(endlist)
```
